## `xor`

Exclusive or  
Either a or b but ***not*** both

````python
a = True
b = True

# A
A_xor = a + b

# B
B_xor = a + b == 1

# C
C_xor = not a != (not b)

print(f"A: {A_xor}")
print(f"B: {B_xor}")
print(f"C: {C_xor}")
````

There is also an operator for `xor`: `^`


# `elif`

We can nest if statements like the following:

````python
day = input("What day of the week is it?")

if day == "Monday":
  print("Start the weekly lab assignment")
else:
  if day == "Tuesday":
    print("Read the textbook")
  else:
    if day == "Wednesday":
      print("Review notes")
    else:
      if day == "Thursday":
        print("Finish Lab")
      else:
        print("I don't know what day that is, but you should probably do a programming exercise.")
````
